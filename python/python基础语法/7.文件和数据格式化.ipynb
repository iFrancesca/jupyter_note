## 7.1.2文件的打开关闭


|文件打开模式|含义|
|--|--|
|'r'|只读模式，如果文件不存在，返回异常FileNotFundError,默认值|
|'w'|覆盖写模式，文件不存在则创建，存在则完全覆盖|
|'x'|创建写模式，文件不存在则创建，存在则返回异常FileExistsError|
|'a'|追加写模式，文件不存在则创建，存在则在文件最后追加内容|
|'b'|二进制文件模式|
|'t'|文本文件模式，默认值|
|'+'|与r/w/a/x一同使用，在原功能基础上增加同时读写功能|

## 7.1.3文件的读写

### 文件内容读取方法
|操作方法|含义|如果给出参数|
|-|-|-|
|<file>.read(size=-1)|从文件中读入整个文件内容|读入前size长度的字符串或字节流|
|<file>.readline(size=-1)|从文件中读入一行内容|读入该行前size长度的字符串或字节流|
|<file>.readlines(hint=-1)|从文件中读入所有行|读入hint行|

````python
path=input('输入打开文件位置：')
f=open(path,'r+',encoding='utf-8')
````

````python
print(f.read())
f.seek(0)
print(f.read(8))
````

### 文件内容写入方法
||操作方法|含义|
|-|-|
|<file>.wite(s)|向文件写入一个字符串或字节流|
|<file>.writelines(lines)|将一个元素全为字符串的列表写入文件|
|<file>.seek(offset)|改变当前文件操作指针的文职，offset的值：0——文件开头，1——当前位置，2——文件结尾|

### 按数量读入，逐步处理

````python
fname=input('输入打开文件位置：')
fo=open(fname,'r',encoding='utf-8')
fo.seek(0)
txt2=fo.read(2)
while txt2!='':
    txt2+=fo.read(2)
print(txt2)
````

### 逐行遍历文本
**1一次读入，分行处理**

````python
fname=input('输入打开文件位置：')
fo=open(fname,'r',encoding='utf-8')
for line in fo.readlines():
    print(line)
fo.close()
````

**2分行读入，逐行处理**

````python
fname=input('输入打开文件位置：')
fo=open(fname,'r',encoding='utf-8')
for line in fo:
    print(line)
fo.close()
````

 # 7.7 json库的使用
### json库概述


>json库是处理JSON格式的Python标准库，主要包含两类函数：**操作类函数**和**解析类函数**

|函数|功能|
|-|-|
|操作类函数|完成外部JSON格式和程序内部的数据类型之间的转换|
|解析类函数|解析键值对内容|

|json格式|Pyson数据类型|
|-|-|
|对象{}|字典|
|数组[]|列表|
|string|str|

### json库的解析
>编码（encoding）&解码（decoding）--->数据类型序列化和反序列化过程
    

> 序列化：将对象数据类型转换为可存储或网络传输格式的过程

* json库的4个操作类函数


|函数|描述|
|-|-|
|json.dumps(obj,sort_keys-False,ident=None)|将python数据类型转换为json格式，编码过程|
|json.loads(string)|json格式-->python数据类型，decoding|
|json.dump(obj,fp,sort_keys=False,indent=None)|与dumps()功能一致，输出到文件fp|
|json.load(string,fp)|与load()功能一致，输出到文件fp

* sort_keys:对字典元素按照Key进行排序
* indent：增加数据缩进

````python
import json
help(json)
````

````python
import json
dt={'b':5,"a":9,"c":1}
````

````python
s1=json.dumps(dt)
s2=json.dumps(dt,sort_keys=True,indent=5)
s3=json.dumps(dt,sort_keys=True,indent=3)
print(s1,'\n',s2,'\n',s3)
````

````python
print(json.loads(s2))
print(json.loads(s1))
````

````python
print(json.dumps("中国"))
print(json.dumps("中国",ensure_ascii=False))     #json.dumps 序列化时对中文默认使用的ascii编码.想输出真正的中文需要指定ensure_ascii=False：
````
