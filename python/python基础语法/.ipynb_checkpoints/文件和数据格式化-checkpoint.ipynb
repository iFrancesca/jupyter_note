### 二维数据写入CSV文件

```python
'''读入二维in.csv文件，将其中数据读出，将数字部分计算百分比后输出到二维out.csv文件中'''
fo=open('C:\\Users\\lenovo\\Desktop\\计算机\\python\\00\\二维in.csv')
fw=open('C:\\Users\\lenovo\\Desktop\\计算机\\python\\00\\二维out.csv','w+')
ls=[]
for line in fo:      #将CSV文件中的二维数据读入到列表变量
    line=line.replace('\n','')
    ls.append(line.split(','))
for line in ls:        #遍历列表变量中的两位数据输出到CSV文件
    for j in range(len(line)):
        if line[j].replace('.','').isnumeric():
            line[j]='{:.2%}'.format(eval(line[j])/100)
for row in ls:         #将列表变量中的两位数据输出到CSV文件
    print(row)
    fw.write(','.join(row)+'\n')
fo.close()
fw.close()
```
