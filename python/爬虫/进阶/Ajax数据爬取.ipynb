```python
import requests
from urllib.parse import urlencode
import os
import os.path as op
from hashlib import md5

def get_page(offset):

    params={
        
        'aid':24,
        'app_name':'web_search',
        'format':'json',
        'offset':offset,
        'autoload':'true',
        'count':20,
        'en_qc':1,
        'cur_tab':1,
        'from':'search_tab',
        'pd':'synthesis'
    }
   
    headers={
        'X-Requested-With':'XMLHttpRequest',
        'Cookies':'tt_webid=6788393831844185614; WEATHER_CITY=%E5%8C%97%E4%BA%AC; tt_webid=6788393831844185614; csrftoken=0fd39b0f026752a8f38a902ab9226d44; s_v_web_id=k64mh7pe_9IueD6zl_X1iG_4smt_8Dkd_SMrb5NXXMR8G; __tasessionId=naf9b8ylp1580623604878',
        'Host':'www.toutiao.com',
        'Referer':'https://www.toutiao.com/search/?keyword=%E8%A1%97%E6%8B%8D',
        'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'
            }
    try:
        url='https://www.toutiao.com/api/search/content/?keyword=街拍'+urlencode(params)
        response=requests.get(url,headers=headers)
        response.raise_for_status
        response.encoding=response.apparent_encoding
        return response.json()
    except:
        print('爬取出错')

def get_images(json):
    if json.get('data'):
        for item in json.get('data'):
            title=item.get('title')
            images=item.get('image_list')
            if images!=None:
                for image in images:
                    yield{
                        'image':image.get('url'),
                        'title':title
                    }
    return images


def save_to_file(item):
    path='d:/python/爬虫/图片/'+item.get('title')
    if not op.exists(path):
        os.mkdir(path)
        
    try:
        response=requests.get(item.get('image'))
        file_path=path+'/{0}.{1}'.format(md5(response.content).hexdigest(),'jpg')
        if not op.exists(file_path):
            with open(file_path,'wb') as f:
                f.write(response.content)
        else:
            print('Already Download')
    except:
        print('Fail to Save Image')

def main(offset):
    json=get_page(offset)
    for item in get_images(json):
        save_to_file(item)
        
if __name__=='__main__':
    for i in range(1,5):
        main(i*20)
```
