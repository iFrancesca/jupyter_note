## Pandas操作EXEL的行列

1.读取指定的单行，数据会存在列表里面

```python
data=df.ix[0].values     #0表示第一行 这里读取数据并不包含表头
print(data)
```

2：读取指定的多行，数据会存在嵌套的列表里面

```python
data=df.ix[[5,10]].values  #读取指定多行的话，就要在ix[]里面嵌套列表指定行数
print(data)
```

3：读取指定的行列

```python
data=df.ix[3,1]  #读取第一行第二列的值，这里不需要嵌套列表
print(data)
```

4：读取指定的多行多列值

```python
data=df.ix[[2,5],['任务gps经度','任务标价','任务执行情况']].values
print(data)
```

5：获取所有行的指定列

```python
data=df.ix[:,['任务标价','任务执行情况']]
print(data.values)
print(data)
```

6：获取行号并打印输出

```python
print('输出行号列表',df.index.values)
```

7：获取列名并打印输出

```python
print(df.columns.values)  #列号
```

8：获取指定行数的值

```python
print('输出值',df.sample(3).values)#这个方法类似于head()方法以及df.values方法
                                    # 任意三行的值
    
```

9：获取指定列的值

```python
print(df['任务标价'].values)
```

## Pandas处理Excel数据成为字典

```python
test_data=[]
for i in df.index.values:
    row_data=df.ix[i,['任务号码','任务gps 纬度','任务gps经度','任务标价','任务执行情况']].to_dict()
     #根据i来获取每一行指定的数据 并利用to_dict转成字典
    test_data.append(row_data)
print(test_data)
```
