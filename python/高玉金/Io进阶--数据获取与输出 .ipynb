# 文件格式转换

## csv文件的读取、修改与写入

* 逗号分隔数值的存储格式叫做CSV格式（comma separated values，即逗号分隔值），是一种通用的、相对简单的文件格式，在商业和科学上广泛应用，尤其应用在程序之间转移表格数据
* 纯文本格式，通过单一编码表表示字符
* 以行为单位，开头不留空行，行之间没有空行
* 每行表示一个一维数据，多行表示二位数据
* 以逗号分隔每列数据，列数据为空也要保留逗号
* 可以包含或不包含列名，包含列名放置在文件第一行

#### 手工处理CSV文件
* 使用PYTHON的OPEN函数读入CSV文件
* 使用READLINES()方法全部读入一个列表中，处理行尾的回车符
* 使用字符串的SPLIT()进行切割，得到切割后的列表

```python
with open('D:\\python\\数据处理\\二维in.csv','r') as f:
    ls=[]
    for line in f:
        line=line.strip()
        ls.append(line.split(','))
ls
```

```python
with open('D:\\python\\数据处理\\二维in.csv','r') as f:
    ls=[]
    for line in f:
        line=line.replace('\n','')
        ls.append(line.split(','))
ls
```
