```python
def read_before_file(file_path, line='快优'):
    
    month = re.findall('(\d).*?', file_path)[0] + '月'
    
    if line == '快优':
        data = pd.read_excel(file_path, 'Sheet3').iloc[:10,:11]        
        data.columns = ['大区','荣誉感','服务要求','服务奖励','司机部落','月会宣教',\
                        '司服沟通','绩效考核','总体满意度','得分','前一月得分']
        data.drop(columns='前一月得分', inplace=True)
       
    elif line == 'D1':
        data = pd.read_excel(file_path, 'Sheet3').iloc[:6, :10]
        data.columns = ['大区','荣誉感','服务要求','服务奖励','司机部落','月会宣教',\
                        '司服沟通','绩效考核','总体满意度','得分']


        
    data['月份'] = month 
    # 原始数据的其他更改， 例如列名等
    data = data[data.大区.notna()]

    return data
```

```python
# 生成文件名的列表
yx_path = list(d for d in os.listdir() if '优享' in d)
kc_path = list(d for d in os.listdir() if '快车' in d)
d1_path = list(d for d in os.listdir() if 'D1' in d)
yx_path
```

```python
# 提取文件中的数据
yx_lt = list(map(read_before_file, yx_path))
kc_lt = list(map(read_before_file, kc_path))
d1_lt = list(map(read_before_file, d1_path, ['D1']*len(d1_path) ))

# 将文件的中的数据组合为一个DataFrame
df1_yx = pd.concat(yx_lt, axis=0)
df1_kc = pd.concat(kc_lt, axis=0)
df1_d1 = pd.concat(d1_lt, axis=0)
```

```python
def read_after_pt(file_path):
    
    month = re.findall('(.*)队长对平台.*?.xlsx', file_path)[0]
    df_yx = pd.read_excel(file_path, '优享').iloc[:9]  # A1:J10
    df_kc = pd.read_excel(file_path, '快车').iloc[:9]  # A1:J10
    df_d1 = pd.read_excel(file_path, 'D1').iloc[:5]  # A1:J5
    df_yx.columns = ['大区','荣誉感','服务要求','服务奖励','司机部落','月会宣教',\
                        '司服沟通','绩效考核','总体满意度','得分']
    df_kc.columns = ['大区','荣誉感','服务要求','服务奖励','司机部落','月会宣教',\
                        '司服沟通','绩效考核','总体满意度','得分']
    df_d1.columns = ['大区','荣誉感','服务要求','服务奖励','司机部落','月会宣教',\
                        '司服沟通','绩效考核','总体满意度','得分']
    
    df_yx['月份'] = month
    df_kc['月份'] = month
    df_d1['月份'] = month
    df_yx = df_yx[df_yx.大区.notna()]
    df_kc = df_kc[df_kc.大区.notna()]
    df_d1 = df_d1[df_d1.大区.notna()]

    return df_yx, df_kc, df_d1
```

```python
import matplotlib.pyplot as plt

region = ['鲁津冀区域', '华东区域', '华北区域', '中南区域', '中西区域', '沪苏区域', '西南区域', '广东区域','全国']
region_d1 = ['华东区域', '广东区域', '中南区域', '全国']
Q = ['荣誉感', '服务要求', '服务奖励', '司机部落', '月会宣教', '司服沟通', '绩效考核', '总体满意度','得分']

def plot_df(dfs, line='快优'):
    global region, region_d1, Q
    if line == '快优':
        for reg in region:
            df = dfs.query('大区 == @reg')
            df.sort_values(by=['月份'])
            for y in q:
                plt.plot(df['月份'], df[q[y]])
            
    elif line == 'D1':
        for reg in region_d1:
            df = dfs.query('大区 == @reg')
            df.sort_values(by=['月份'], inplace=True)
            for q in Q:
                plt.plot(df['月份'], df[q])
    return 'ok'
```
