```python
def num_per(df):
    global item
    s = df[item].value_counts()
    d = pd.DataFrame( {'value':s.index.values, 'num': s.values, 'per': s.values/s.cumsum().iloc[-1]} )
    d.sort_values(by=['value'], inplace=True)
    d['per_cum'] = d['per'].cumsum()
    return d

d = num_per(df)
'''
d_yx = num_per(df_yx)
d_kc = num_per(df_kc)
d_d1 = num_per(df_d1)
'''
pd.set_option('display.max_rows',None)
d  # 查看最后几行的数据: df.tail()
```

```python
def des_num_per(df):
    global item
    s = df[item].value_counts()
    d = pd.DataFrame( {'value':s.index.values, 'num': s.values, 'per': s.values/s.cumsum().iloc[-1]} )
    d.sort_values(by=['value'], ascending=False, inplace=True)
    d['per_cum'] = d['per'].cumsum()
    return d

d2 = des_num_per(df)
'''
d2_yx = des_num_per(df_yx)
d2_kc = des_num_per(df_kc)
d2_d1 = des_num_per(df_d1)
'''
d2  # 查看最后几行的数据
```

## 得分分组计算

```python
# bins=[0, n1, n2, n3, n4, n5, df[item].describe()['max']+0.1]
bins=[0, 44, 66.7, 82, 99, df[item].describe()['max']+0.1]
fenzu = pd.cut(df[item], bins, right=False)
#sort_index按区间顺序排序
s = fenzu.value_counts()
df_fenzu = pd.DataFrame( {item:s.index.values, 'num': s.values, 'per': s.values/s.cumsum().iloc[-1]} )
df_fenzu.set_index([item], drop=True, inplace=True)
df_fenzu.sort_index(ascending=True, inplace=True)
df_fenzu['per_cum'] = df_fenzu['per'].cumsum()
df_fenzu[['per', 'per_cum']] = df_fenzu[['per', 'per_cum']].applymap(lambda x: format(x, '.2%'))
df_fenzu
```
