## 覆盖率--准出进度

```python
path1 = '1217/邮件.xlsx'
path2 = '1214/邮件.xlsx'
path3 = '1210/邮件.xlsx'
path4 = '1207/邮件.xlsx'
```

```python
path = []
path.append(path1)
path.append(path2)
path.append(path3)
```

```python
fg1 = pd.read_excel(path1,'1-队长覆盖率')
fg2 = pd.read_excel(path2,'1-队长覆盖率')
fg3 = pd.read_excel(path3,'1-队长覆盖率')
fg4 = pd.read_excel(path4,'1-队长覆盖率')
zc1 = pd.read_excel(path1,'2-准出进度')
zc2 = pd.read_excel(path2,'2-准出进度')
zc3 = pd.read_excel(path3,'2-准出进度')
zc4 = pd.read_excel(path4,'2-准出进度')
```

```python
fg1 = fg1.add_suffix('_1130')
fg2 = fg2.add_suffix('_1126')
fg3 = fg3.add_suffix('_1123')
fg4 = fg4.add_suffix('_1119')
fg1.rename(columns={'大区_1130':'区域'},inplace=True)
fg2.rename(columns={'大区_1126':'区域'},inplace=True)
fg3.rename(columns={'大区_1123':'区域'},inplace=True)
fg4.rename(columns={'大区_1119':'区域'},inplace=True)
fg = pd.merge(fg1,fg2,on='区域')
fg = pd.merge(fg,fg3,on='区域')
fg = pd.merge(fg,fg4,on='区域')
fg
```

```python
columns = ['区域','红线封禁需准出',\
                    '质量不合格需准出_快车','质量不合格需准出_优享','质量不合格需准出_总计','质量不合格需准出_总计（新）',\
                    '质量不合格已准出_快车','质量不合格已准出_优享','质量不合格已准出_总计',\
                    '不合格_准出进度',\
                    '观察期_总计','观察期_已准出','剔除观察期_准出进度',\
                    '强准出_总计','强准出_已准出','强准出_准出进度',\
                    '非磐石目标准出','非目标_已准出','非目标准出进度']
zc1.set_axis(columns,axis=1,inplace=True)
zc1 = zc1.drop(index=[0])

zc2.set_axis(columns,axis=1,inplace=True)
zc2 = zc2.drop(index=[0])

zc3.set_axis(columns,axis=1,inplace=True)
zc3 = zc3.drop(index=[0])

zc4.set_axis(columns,axis=1,inplace=True)
zc4 = zc4.drop(index=[0])
```

```python
zc1 = zc1.add_suffix('_1130')
zc2 = zc2.add_suffix('_1126')
zc1.rename(columns={'区域_1130':'区域'},inplace=True)
zc2.rename(columns={'区域_1126':'区域'},inplace=True)
zc3 = zc3.add_suffix('_1123')
zc4 = zc4.add_suffix('_1119')
zc3.rename(columns={'区域_1123':'区域'},inplace=True)
zc4.rename(columns={'区域_1119':'区域'},inplace=True)
zc = pd.merge(zc1,zc2,on='区域')
zc = pd.merge(zc,zc3,on='区域')
zc = pd.merge(zc,zc4,on='区域')
zc
```

```python
lt_sort = ['中西区域', '中南区域', '西南区域', '鲁津冀区域', '华东区域', '华北区域', '沪苏区域', '广东区域','全国']
# 覆盖 排序
fg['区域'] = fg['区域'].astype('category').cat.set_categories(lt_sort)
fg = fg.sort_values(by=['区域'])
# 准出 排序
zc['区域'] = zc['区域'].astype('category').cat.set_categories(lt_sort)
zc = zc.sort_values(by=['区域'])
```

```python
df = pd.DataFrame({'区域':fg['区域'], '整体覆盖率_1130':fg['整体覆盖率_1130'], '整体覆盖率_1126':fg['整体覆盖率_1126'],\
                   '整体覆盖率_1123':fg['整体覆盖率_1123'], '整体覆盖率_1119':fg['整体覆盖率_1119'],\
                  '剔除观察期_1130':zc['剔除观察期_准出进度_1130'],'剔除观察期_1126':zc['剔除观察期_准出进度_1126'],\
                  '剔除观察期_1123':zc['剔除观察期_准出进度_1123'],'剔除观察期_1119':zc['剔除观察期_准出进度_1119']})
df['整体覆盖率-变化1130'] = df['整体覆盖率_1130'] - df['整体覆盖率_1126']
df['整体覆盖率-变化1126'] = df['整体覆盖率_1126'] - df['整体覆盖率_1123']
df['整体覆盖率-变化1123'] = df['整体覆盖率_1123'] - df['整体覆盖率_1119']
df['准出进度-变化1130'] = df['剔除观察期_1130'] - df['剔除观察期_1126']
df['准出进度-变化1126'] = df['剔除观察期_1126'] - df['剔除观察期_1123']
df['准出进度-变化1123'] = df['剔除观察期_1123'] - df['剔除观察期_1119']
df
```

```python
writer = pd.ExcelWriter('fx.xlsx')
df.to_excel(writer,'1130',index=False)
writer.save()
```
